# Accuracy histograms

Compute accuracy histograms for verification and classification conditions in test and train sets for 1,3,5,7 shots per model. Histograms show the accuracy results for 30 runs of the model for each shot value.

# Import libraries

```python
import numpy as np
import seaborn as sns
import pandas as pd
import pickle
import scipy.stats
import matplotlib.pyplot as plt 
import scipy.stats as st
import statistics
```

# Choose directory

Available directories correspond to the following models
* SiameseZero: Siamese neural network with no weights. Basically a KNN with euclidean distance.
* SiameseFCN: Siamese fully connected neural network.
* SiameseConvolutional: Siamese convolutional neural network
* SiameseLSTM: Siamese LSTM neural network
* SiameseBiLSTM: Siamese Bidirectional LSTM neural network

For each directory, there might be the following subdirectories (not all models have both):
* AllSupport: Contains results using as support the class average for all phrases
* TrainSupport: Contains results using as support the class average of only train phrases

```python
model = "SiameseZero"
subdir = "TrainSupport"
directory = model + "/" + subdir + "/"
```

# Confidence interval function

```python
def mean_confidence_interval(data, confidence=0.95):
    alpha = 1-confidence
    mean = np.mean(data)
    n = len(data)
    S = statistics.stdev(data)
    SE = S/np.sqrt(n)
    z = st.norm.ppf(round(alpha/2,7))
    return round(mean + z*SE,4), round(mean,4), round(mean - z*SE,4)
```

# Plot

```python
all_shots = [1,3,5,7]
```

```python
#x is an array containing tuples of the form: 
#(v_acc_tr,v_acc_te,c_acc_tr,c_acc_te,acc_c_class_train,acc_c_class_test,history)

import matplotlib.ticker as ticker
for shots in all_shots:
    with open(directory + 'k' + str(shots) + '.pickle', 'rb') as f:
        x = pickle.load(f)
    #X contains v_acc_tr,v_acc_te,c_acc_tr,c_acc_te,acc_c_class_train,acc_c_class_test,history
    v_acc_tr = [a[0] for a in x] #Verification accuracy training
    v_acc_te = [a[1] for a in x] #Verification accuracy test
    c_acc_tr = [a[2] for a in x] #Classification accuracy training
    c_acc_te = [a[3] for a in x] #Classificatiofsin accuracy test
    acc_c_class_tr = [a[4] for a in x] #Classification accuracy per class training
    acc_c_class_te = [a[5] for a in x] #Classification accuracy per class test
    
    #Plot
    fig = plt.figure(figsize=(10, 10))
    plot_name = str(shots)+'-Shot '+ model
    fig.suptitle(plot_name , fontsize=16)
    col_labels = ["v_acc","c_acc"]
    row_labels = ["train","test"]
    data = [v_acc_tr,c_acc_tr,v_acc_te,c_acc_te]
    for i in range(4):
        ax= fig.add_subplot(2, 2, i+1)
        # show mean in position text relative to Axes
        ax.text(0.5, 0.9, 'mean: ' + str(round(np.mean(data[i]),4)),ha='right', va='top',transform=ax.transAxes)
        
        #Fix ticks formatting
        if(model == "SiameseBiLSTM" or model=="SiameseLSTM"):
            if(i==0) and (shots == 1):
                ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
        
        if(i%2==3):
            ax.set_xticks(np.arange(0, 1.0, step=0.01))
        if(i%2==0):
            ax.set_ylabel(row_labels[max(0,i-1)])
        if(i<2):
            ax.set_title(col_labels[i])
        sns.distplot(data[i],bins=5,kde=False,ax=ax)
    plt.show()
    
    #Save figures (optional)
    #fig.savefig(str(shots)+ '_' +model +'.png')
    
    #Show confidence interval for test classification accuracy
    print("Confidence interval for test classification accuracy for " + model)
    print(mean_confidence_interval(c_acc_te))
```
